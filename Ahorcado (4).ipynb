```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0, 0.1, 0.2, 0.5,0.6], [1, -1, 0, 3, -1])
```

```python
figure, axes = plt.subplots()
draw_circle = plt.Circle((0.5, 0.5), 0.3)

axes.set_aspect(1)
axes.add_artist(draw_circle)
plt.title('Circle')
plt.show()
```
